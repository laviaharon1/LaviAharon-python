### Part 1: What is a Loop?
A loop repeats code multiple times. The while loop continues as long as a condition is True.  
Syntax:

```python
while condition:
    # code to repeat
    # this code runs while condition is True
```

Code cell:


```python
# Simple while loop
count = 1

while count <= 5:
    print(f"Count is: {count}")
    count = count + 1

print("Loop finished!")
```

--- 
### Part 2: Loop with User Input
Code cell:

```python
# Keep asking until user types "stop"
while True:
    answer = input("Type 'stop' to exit: ")
    
    if answer == "stop":
        break
    
    print("You typed:", answer)

print("Program ended")
```

---
### Part 3: The break Statement
The break statement exits the loop immediately.  
Code cell:

```python
# Find first number divisible by 7
number = 1

while number <= 100:
    if number % 7 == 0:
        print(f"First number divisible by 7 is: {number}")
        break
    number = number + 1
```

---
### Part 4: The continue Statement
The continue statement skips to the next iteration.   
Code cell:

```python
# Print only odd numbers from 1 to 10
number = 0

while number < 10:
    number = number + 1
    
    if number % 2 == 0:
        continue  # Skip even numbers
    
    print(number)
```

---
### Part 5: Common Patterns
Pattern 1: Counter

```python
count = 0
while count < 5:
    print("Hello")
    count = count + 1
```

Pattern 2: User Input Validation


```python
age = -1
while age < 0:
    age = int(input("Enter your age: "))
    if age < 0:
        print("Age cannot be negative!")
        continue
    print(f"Your age is: {age}")
```

Pattern 3: Menu System


```python
choice = ""
while choice != "quit":
    print("1. Option A")
    print("2. Option B")
    print("Type 'quit' to exit")
    choice = input("Choose: ")
```

--- 
### Important Notes

- The condition is checked before each iteration
- Make sure the condition will eventually become False
- Always update the variable in the condition to avoid infinite loops
- Use break to exit early
- Use continue to skip to next iteration

- User keeps guessing

### Task 5: Menu System
Create a simple calculator menu:

- Show options: 1-Add, 2-Subtract, 3-Multiply, 4-Divide, 5-Exit
- Ask user to choose
- If 1-4: ask for two numbers and show result
- If 5: exit the program
- Keep showing menu until user exits

Code cell:

```python
# Your code here

while True:
    option = input('1-Add, 2-Subtract, 3-Multiply, 4-Divide, 5-Exit: ')
    if option == '5':
        print("Exiting.")
        break
    else:
        first = int(input("Enter first number: "))
        second = int(input("Enter second number: "))
        
        if option == '1':
            print(f"Result: {first + second}")
        elif option == '2':
            print(f"Result: {first - second}")
        elif option == '3':
            print(f"Result: {first * second}")
        elif option == '4':
            if second != 0:
                print(f"Result: {first / second}")
            else:
                print("Cannot divide by zero.")
```

Write a program that:

- Asks the user to enter numbers
- Only accepts even numbers
- If odd, print "Only even numbers!" and ask again

### Advanced Challenge (Optional)
Create a simple ATM simulation:

1. Start with balance of 1000
2. Menu options:

    - Check balance
    - Deposit money
    - Withdraw money (check if enough balance)
    - Exit


3. Keep running until user chooses exit
4. Validate all inputs (no negative numbers)

Code cell:

```python
# Your challenge code here
balance = 1000
while True:
    action = input("1-Check Balance, 2-Deposit, 3-Withdraw, 4-Exit: ")
    if action == '1':
        print(f"Your balance is: ${balance}")
    elif action == '2':
        amount = float(input("Enter amount to deposit: "))
        balance = balance + amount
    elif action == '3':
        amount = float(input("Enter amount to withdraw: "))
        if amount > balance:
            print("You don't have enough money!")
            continue
        balance = balance - amount
    elif action == '4':
        print(f"Exiting. Final balance: ${balance}")
        break
    else:
        print("Invalid action. Please try again.")
```

Code cell for homework:

```python
# Multipication table

num = int(input("Enter a number: "))

multiplier = 1
while multiplier <= 10:
    print(f"{num} x {multiplier} = {num * multiplier}")
    multiplier += 1
```

```python
# Positive Numbers Only

total = 0
count = 0

while True:
    num = int(input("Enter a positive number (-1 to stop): "))

    if num == -1:
        break
    if num > 0:
        total += num
        count += 1

if count > 0:
    print(f"Average: {total / count}")
else:
    print("No positive numbers entered")
```

```python
# Simple Game

points = 100

while points > 0:
    action = input("Type 'lose' to lose points or 'exit' to exit: ")
    if action == 'lose':
        points -= 10
        print(f"You lost 10 points. Remaining points: {points}")
    elif action == 'exit':
        print(f"Exiting the game with {points} points.")
        break
    else:
        print("Invalid action. Please type 'hit' or 'exit'.")
```
